```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python

# Load dataset (update path if needed)
df = pd.read_csv("loan_large_dataset.csv")
df.head()
```

## Basic Dataset Info

```python

df.info()
df.describe()
```

## Loan Status Distribution

```python

sns.countplot(x="Loan_Status", data=df)
plt.title("Loan Status Distribution")
plt.show()
```

## Credit History vs Loan Status

```python

sns.countplot(x="Credit_History", hue="Loan_Status", data=df)
plt.title("Credit History vs Loan Status")
plt.show()
```

## Income Distribution by Loan Status

```python

sns.histplot(data=df, x="ApplicantIncome", hue="Loan_Status", kde=True, bins=30)
plt.title("Income Distribution by Loan Status")
plt.show()
```

## Education vs Loan Status

```python

sns.countplot(x="Education", hue="Loan_Status", data=df)
plt.title("Education vs Loan Status")
plt.show()
```

## Property Area vs Loan Status

```python

sns.countplot(x="Property_Area", hue="Loan_Status", data=df)
plt.title("Property Area vs Loan Status")
plt.show()
```

## Approval Rate by Gender

```python

gender_rate = (
    df.groupby("Gender")["Loan_Status"]
    .apply(lambda x: (x == "Y").mean() * 100)
    .reset_index(name="Approval Rate (%)")
)

sns.barplot(x="Gender", y="Approval Rate (%)", data=gender_rate)
plt.ylim(0, 100)
plt.title("Approval Rate by Gender")
plt.show()
```

## Correlation Heatmap

```python

corr = df.select_dtypes(include=np.number).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```


## Key Insights
- Credit history strongly affects loan approval  
- Income distribution varies by approval status  
- Gender and education show noticeable approval differences
